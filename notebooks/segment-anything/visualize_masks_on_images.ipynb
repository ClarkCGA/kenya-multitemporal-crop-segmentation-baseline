```python
import numpy as np
import rasterio
import xarray as xr
import matplotlib.pyplot as plt
import os
from pathlib import Path
import torch
```

```python
imagedir = Path('/home/data/kenya-sam/images/')
samdir = Path('/home/data/kenya-sam/labels/')
maskdir = Path('/home/data/kenya/sam-masks/')
```

```python
mask_paths = [Path(maskdir).joinpath(i) for i in os.listdir(maskdir) if i.endswith('.tif')]
image_paths = [Path(imagedir).joinpath(i) for i in os.listdir(maskdir) if i.endswith('.tif')]
sam_paths = [Path(samdir).joinpath(i) for i in os.listdir(maskdir) if i.endswith('.tif')]
```

```python
def plot_images(i):
    image_path, mask_path, sam_path = image_paths[i], mask_paths[i], sam_paths[i]
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))

    with rasterio.open(image_path) as src:
        image = src.read().transpose(1,2,0)[:,:,:3]
    with rasterio.open(mask_path) as src:
        mask = src.read().squeeze(0)
    with rasterio.open(sam_path) as src:
        sammask = src.read().squeeze(0)
    print(image_path)
    axes[0].imshow(image)
    axes[0].imshow(mask, cmap='viridis', alpha=0.3)
    axes[1].imshow(image)
    axes[1].imshow(sammask, cmap='viridis', alpha=0.3)
```

```python
plot_images(203)
```
